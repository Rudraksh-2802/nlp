Task 1 : Conceptual Understanding

1.Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different words because models are case-sensitive. This can create duplicate entries and increase data inconsistency. To avoid this, text is usually converted to lowercase during preprocessing.

2.What happens if stopwords are not removed ?
If stopwords are not removed, the dataset gets filled with common words like "is", "the", and "and". These words add little meaning but increase noise and processing time. This can reduce the overall performance of the model.

3.Two real-world scenarios where removing stopwords can be harmful
In sentiment analysis, removing words like "not" can completely change the meaning of a sentence. In question-answering systems, removing words like "what" or "how" can make the sentence unclear. So, stopword removal must be done carefully based on context.

4. Difference between Stemming and Lemmatization
Stemming reduces words to their root form by cutting parts of the word, which may not always be meaningful. Lemmatization converts words into their correct base form using grammar rules. Therefore, lemmatization is more accurate but slightly slower than stemming.

```python
# Task 2 : Build Advanced Preprocessing Function

import re

def preprocess_text(text):
    # handles empty or invalid input
    if not text or not isinstance(text, str):
        return [], ""
    # removes urls and emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)
    # removes numbers
    text = re.sub(r"\d+", "", text)
    # converts to lowercase
    text = text.lower()
    # removes repeated characters (soooo -> so)
    text = re.sub(r"(.)\1{1,}", r"\1", text)
    # removes special characters
    text = re.sub(r"[^a-z\s]", "", text)
    # removes extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    # tokenize
    words = text.split()
    # removes short words except 'no' and 'not'
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]
    cleaned_text = " ".join(words)
    return words, cleaned_text
```

```python
# Task 3 : Stress Testing

test_sentences = [
"Get 100% FREE access now!!!",
"I absolutely looooved this product 😍😍",
"Worst service ever... 0/10",
"Call me at 9876543210",
"This is THE best course!!!",
"Visit https://openai.com now!",
"Nooooo this is baaad!!!",
"OK OK OK I got it",
"Win $$$ now!!! Limited offer!!!",
"I am not happy with this",
"Broooo this movie is littt 🔥🔥",
"Email me at test@gmail.com asap!!!"
]

for text in test_sentences:
    tokens, cleaned = preprocess_text(text)
    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("=" * 60)
```

```python
# Task 4 : Token Analytics

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    if tokens:
        total = len(tokens)
        unique = len(set(tokens))
        avg_len = sum(len(w) for w in tokens) / total
        print("Sentence:", text)
        print("Total Tokens:", total)
        print("Unique Tokens:", unique)
        print("Average Token Length:", round(avg_len, 2))
        print("=" * 60)
```

Task 4 : Token Analysis 

Analysis questions :

1)The sentence "Get 100% FREE access now!!!" had the most noise. It contains numbers, uppercase words, and multiple special characters. These elements make the text messy and require several preprocessing steps to clean.
2)The sentence "I am not happy with this" retained the most meaningful tokens after cleaning. Important words like "not" and "happy" were preserved. This helps maintain the original sentiment and meaning of the sentence.

```python
# Task 5 : Frequency Analysis

from collections import Counter

all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

```python
# Task 6 : Build Full Pipeline

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)

print(result)
```
